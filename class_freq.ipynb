```python
import mmcv
```

```python
class_split = mmcv.load('appendix/mured/class_split.pkl')  
```

```python
class_freq = mmcv.load('appendix/mured/class_freq.pkl')
class_freq
```

```python
new_condition_prob = class_freq['condition_prob'][:, class_split['mid']]
new_condition_prob = new_condition_prob[class_split['mid'],:]
new_condition_prob

class_freq['condition_prob'] = new_condition_prob
```

```python
import numpy as np
new_gt_labels = np.array(class_freq['gt_labels'])[:, class_split['mid']]
class_freq['gt_labels'] = new_gt_labels
class_freq['gt_labels']
```

```python
class_freq['neg_class_freq'] = class_freq['neg_class_freq'][class_split['mid']]
class_freq['neg_class_freq']
```

```python
class_freq['class_freq'] = class_freq['class_freq'][class_split['mid']]
class_freq['class_freq']
```

```python
class_freq['gt_labels']
```

```python
mmcv.dump(class_freq, 'appendix/mured/group2/class_freq_new.pkl')
```
